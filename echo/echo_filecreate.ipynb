```python
import pandas as pd
import os


csv_path = 'D:/mimiciv_data/daiv_preprocessed/MI_M.csv'
df = pd.read_csv(csv_path)

base_download_path = 'D:/mimiciv_data/mimic-iv-echo/0.1/files/MI_M_echo'

for index, row in df.iterrows():
    subject_id = row['subject_id']
    echo_study_id = row['echo_study_id']
    
    
    subject_folder = f"p{str(subject_id)[:2]}/p{subject_id}"
    study_folder = f"s{echo_study_id}"
    download_path = os.path.join(base_download_path, subject_folder, study_folder)

    if not os.path.exists(download_path):
        os.makedirs(download_path) 
        print(f"{download_path} 경로에 폴더가 생성되었습니다.")
```
