```python
import pandas as pd
import os


csv_path = 'D:/mimiciv_data/daiv_preprocessed/MI_M.csv'
df = pd.read_csv(csv_path)

base_download_path = 'D:/mimiciv_data/mimic-iv-echo/0.1/files/MI_M_echo1'

for index, row in df.iterrows():
    subject_id = row['subject_id']
    echo_study_id = row['echo_study_id']
    
    
    subject_folder = f"p{str(subject_id)[:2]}/p{subject_id}"
    study_folder = f"s{echo_study_id}"
    download_path = os.path.join(base_download_path, subject_folder, study_folder)

    if not os.path.exists(download_path):
        os.makedirs(download_path) 
        print(f"{download_path} 경로에 폴더가 생성되었습니다.")
```

```python
records = []
notrecords = []

for label in ['MI_M_echo']:
    label_dir = os.path.join('D:/mimiciv_data/mimic-iv-echo/0.1/files', label)
    for patient in os.listdir(label_dir):
        patient_dir = os.path.join(label_dir, patient)
        for session in os.listdir(patient_dir):
            session_dir = os.path.join(patient_dir, session)
            dcm_files_found = False  # .dcm 파일 발견 여부를 추적하기 위한 플래그
            for file in os.listdir(session_dir):
                if file.endswith('.dcm'):
                    dcm_files_found = True  # .dcm 파일이 하나라도 발견되면 플래그를 True로 설정
                    record_path = os.path.join(session_dir, file[:-4])  # '.dcm' 확장자 제거
                    records.append((record_path, 1))
            if not dcm_files_found:  # 현재 세션 디렉토리에서 .dcm 파일이 발견되지 않으면
                print(f"No .dcm files found in: {session_dir}")  # 해당 경로 출력


print(records)
```
