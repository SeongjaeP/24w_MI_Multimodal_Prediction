```python
import os
import pandas as pd

# DataFrame에서 정보 로드
folder_path = 'D:\\physionet.org\\files\\mimic-iv-ecg\\1.0\\files'
MI_F_list = pd.read_csv('C:\\Users\\Administrator\\24w_MI_Multimodal_Prediction\\processed_data\\MI_F.csv')

subjects = MI_F_list['subject_id']
ecg_studies = MI_F_list['ecg_study_id']

ecg_paths = []

for i in range(len(subjects)):
    # 파일 경로 구성
    p = 'p' + str(subjects[i])[:4]
    sub = 'p' + str(subjects[i])
    ecg_study = 's' + str(ecg_studies[i])
    
    # .dat 및 .hea 파일 경로
    dat_file_path = os.path.join(folder_path, p, sub, ecg_study, f"{ecg_studies[i]}.dat")
    hea_file_path = os.path.join(folder_path, p, sub, ecg_study, f"{ecg_studies[i]}.hea")
    
    # 파일 존재 여부 확인
    if os.path.exists(dat_file_path) and os.path.exists(hea_file_path):
        # 파일 경로 저장
        ecg_paths.append((dat_file_path, hea_file_path))
    else:
        print(f"Files not found for subject ID {subjects[i]} and study ID {ecg_studies[i]}.")

# ecg_paths에 저장된 경로를 사용하여 추가 작업을 수행할 수 있습니다.
# 예: 각 파일을 읽어서 데이터 처리
```

```python
import requests
from bs4 import BeautifulSoup
import os


base_url = 'https://physionet.org/content/mimic-iv-ecg/1.0/files/p1000/p10000032/s40689238/'
html_file_path = 'D:/physionet.org/files/mimic-iv-ecg/1.0/files/p1000/p10000032/s40689238/index.html'

# HTML 파일 읽기
with open(html_file_path, 'r') as file:
    html_content = file.read()

# BeautifulSoup을 이용해 HTML 파싱
soup = BeautifulSoup(html_content, 'html.parser')

# .dat 및 .hea 파일 링크 찾기
files_to_download = []
for link in soup.find_all('a'):
    href = link.get('href')
    if href and (href.endswith('.dat') or href.endswith('.hea')):
        files_to_download.append(href)

# 파일 다운로드 경로 설정
download_path = 'D:/physionet.org/files/mimic-iv-ecg/1.0/files/p1000/p10000032/s40689238'

# 다운로드할 파일이 있는 경우
if not os.path.exists(download_path):
    os.makedirs(download_path)

# 파일 다운로드
for file in files_to_download:
    file_url = os.path.join(base_url, file)  # 상대 경로를 전체 URL로 변환
    response = requests.get(file_url)
    with open(os.path.join(download_path, file.split('/')[-1]), 'wb') as f:
        f.write(response.content)
    print(f"{file} 다운로드 완료.")
```

```python
import pandas as pd
import requests
import os

# CSV 파일 로드
csv_path = 'C:\\Users\Administrator\\24w_MI_Multimodal_Prediction\\processed_data\\MI_F.csv'
df = pd.read_csv(csv_path)

# 파일이 저장될 기본 경로 설정
base_download_path = 'D:/physionet.org/files/mimic-iv-ecg/1.0/files'

for index, row in df.iterrows():
    subject_id = row['subject_id']
    ecg_study_id = row['ecg_study_id']
    
    # 파일 다운로드 URL과 저장 경로 생성
    subject_folder = f"p{str(subject_id)[:4]}/p{subject_id}"
    study_folder = f"s{ecg_study_id}"
    file_base_url = f"https://physionet.org/content/mimic-iv-ecg/1.0/files/{subject_folder}/{study_folder}/"
    download_path = os.path.join(base_download_path, subject_folder, study_folder)
    
    # 다운로드할 파일이 있는 경우
    if not os.path.exists(download_path):
        os.makedirs(download_path)
    
    try:
        # .dat 및 .hea 파일 다운로드
        for file_ext in ['.dat', '.hea']:
            file_url = f"{file_base_url}{ecg_study_id}{file_ext}"
            response = requests.get(file_url)
            # 요청이 성공했을 경우에만 파일 저장
            if response.status_code == 200:
                with open(os.path.join(download_path, f"{ecg_study_id}{file_ext}"), 'wb') as f:
                    f.write(response.content)
                print(f"{ecg_study_id}{file_ext} 다운로드 완료.")
            else:
                print(f"{ecg_study_id}{file_ext} 파일을 찾을 수 없습니다.")
    except Exception as e:
        print(f"다운로드 중 오류 발생: {e}")
```

```python
from pathlib import Path
import pandas as pd

import wfdb

# get the path to all the study .hea files for p10023771
paths = list(Path("p10023771/.").rglob("*.hea"))

# get date and time for each study
date_times = {'study':[],'date':[],'time':[]} # use a dictionary to store the date and time for each study
for file in paths:
    study = file.stem
    metadata = wfdb.rdheader(f'{file.parent}/{file.stem}')
    date_times['study'].append(study)
    date_times['date'].append(metadata.base_date)
    date_times['time'].append(metadata.base_time)

df_date_times = pd.DataFrame(data=date_times)
df_date_times.to_csv('p10023771_date_times.csv', index=False)
```

```python
import os
import pandas as pd
import re
import requests

# CSV 파일 경로
csv_path = 'C:/Users/Administrator/24w_MI_Multimodal_Prediction/processed_data/NOT_MI_F.csv'

# 기본 다운로드 경로 설정
base_download_path = 'C:/Users/Administrator/24w_MI_Multimodal_Prediction/processed_data/ecg_data/NOT_MI_F_ecg'

# CSV 파일 읽기
df = pd.read_csv(csv_path)

for index, row in df.iterrows():
    subject_id = row['subject_id']
    ecg_study_id = row['ecg_study_id']
    
    # 파일 다운로드 URL 및 저장 경로 생성
    subject_folder = f"p{str(subject_id)[:4]}/p{subject_id}"
    study_folder = f"s{ecg_study_id}"
    file_base_url = f"https://physionet.org/content/mimic-iv-ecg/1.0/files/{subject_folder}/{study_folder}/"
    download_path = os.path.join(base_download_path, subject_folder, study_folder)
    
    # 디렉토리 생성
    if not os.path.exists(download_path):
        os.makedirs(download_path)
    
    # .hea 파일 다운로드 및 메타데이터 추출
    hea_url = f"{file_base_url}{ecg_study_id}.hea"
    response = requests.get(hea_url)
    content = response.text
    
    pattern = fr"({ecg_study_id}\s+12.*?V6)"
    matches = re.findall(pattern, content, re.DOTALL)
    
    # 새로운 .hea 파일 경로
    new_hea_file_path = os.path.join(download_path, f"{ecg_study_id}.hea")
    
    # 파일이 이미 존재하는 경우, 삭제
    if os.path.exists(new_hea_file_path):
        os.remove(new_hea_file_path)
        print(f"Existing file {ecg_study_id}.hea has been removed.")
    
    # 메타데이터가 발견된 경우, 새 파일 저장
    if matches:
        new_hea_content = matches[0].strip()
        with open(new_hea_file_path, 'w') as new_file:
            new_file.write(new_hea_content)
        print(f"Metadata extracted and saved to: {new_hea_file_path}")
    else:
        print(f"No metadata found for {ecg_study_id}.hea")
```
